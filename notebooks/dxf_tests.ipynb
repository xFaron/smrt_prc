```python
import os
import subprocess
from pathlib import Path


def convert_to_dxf(input_path, output_path):
    ext = Path(input_path).suffix.lower()

    if ext == ".dwg":
        subprocess.run(
            [
                "ODAFileConverter",
                str(Path(input_path).parent),
                str(Path(output_path).parent),
                "ACAD2000",
                "DXF",
                "0",
            ]
        )
    # elif ext == ".svg":
    # 	subprocess.run(["inkscape", input_path, "--export-filename=" + output_path])
    # elif ext == ".pdf":
    # 	subprocess.run(["pstoedit", "-f", "dxf:-polyaslines", input_path, output_path])
    # elif ext == ".shp":
    # 	subprocess.run(["ogr2ogr", "-f", "DXF", output_path, input_path])
    # elif ext == ".ai":
    # 	subprocess.run(["inkscape", input_path, "--export-filename=" + output_path])
    elif ext == ".dxf":
        subprocess.run(["cp", input_path, output_path])
    else:
        print(f"Unsupported format: {ext}")
```

```python
convert_to_dxf("input/plan.dxf", "output/plan_converted.dxf")
```

```python
import ezdxf
import matplotlib.pyplot as plt

# Load DXF file
doc = ezdxf.readfile("output/plan_converted.dxf")
msp = doc.modelspace()

# Lists to store geometry
lines = []
circles = []
texts = []

# Extract entities
for entity in msp:
    if entity.dxftype() == "LINE":
        lines.append(
            (
                (entity.dxf.start.x, entity.dxf.start.y),
                (entity.dxf.end.x, entity.dxf.end.y),
            )
        )
    elif entity.dxftype() == "CIRCLE":
        circles.append((entity.dxf.center.x, entity.dxf.center.y, entity.dxf.radius))
    elif entity.dxftype() in ("TEXT", "MTEXT"):
        texts.append((entity.dxf.insert.x, entity.dxf.insert.y, entity.dxf.text))


# # Plot the extracted geometry
# fig, ax = plt.subplots()

# # Draw lines
# for start, end in lines:
#     ax.plot([start[0], end[0]], [start[1], end[1]], color='black')

# # Draw circles
# for x, y, r in circles:
#     circle = plt.Circle((x, y), r, fill=False, color='blue')
#     ax.add_artist(circle)

# # Draw text positions
# for x, y, txt in texts:
#     ax.text(x, y, txt, fontsize=8, color='red')

# ax.set_aspect('equal')
# plt.show()
```
